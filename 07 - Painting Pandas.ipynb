Define groups
---

* Lets group by University (NL, DE, AU)
* We can do multiple grouping

```python
groupGrades = df.groupby('University')
```

```python
#df['Lower'] = groupGrades['Grade'].transform( lambda x: x.quantile(q=.25) - (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )
#df['Upper'] = groupGrades['Grade'].transform( lambda x: x.quantile(q=.75) + (1.5*x.quantile(q=.75)-x.quantile(q=.25)) )
#df['Outlier'] = (df['Grade'] < df['Lower']) | (df['Grade'] > df['Upper'])

df['Low'] = groupGrades['Grade'].transform( lambda x: df['Grade'].mean()-2*df['Grade'].std() )
df['Up'] = groupGrades['Grade'].transform( lambda x: df['Grade'].mean()+2*df['Grade'].std() )
df['Outlier'] = (df['Grade'] < df['Low']) | (df['Grade'] > df['Up'])
df
```

```python
df.plot(style='-')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, sharey=True)
for i, (group, data) in enumerate(groupGrades):
    #ax.get_xaxis().set_visible(False) # hide ticks
    ax[0,i].set_visible(False) # hide ticks
    data.plot( ax=ax[0,i], style='-', table=True, secondary_y=['Outlier'], mark_right=False )
```
