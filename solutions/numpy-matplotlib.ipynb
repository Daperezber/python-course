$\S$ Exercise: Prime number sieve (solution)
===

* Use `ones` to initialize a boolean array with `True` values.
* Assign to all higher multiples of `i` by slicing with `[2*i::i]`.

```python
is_prime = np.ones((100,), dtype=bool)
is_prime[:2] = False

max_factor = int(np.sqrt(len(is_prime)))
for i in range(2, max_factor):
    is_prime[2*i::i] = False

np.nonzero(is_prime)
```

$\S$ Exercise: Scatter plots (solution)
===

The trick to mimic the graphic is to set `xlim` and `ylim`.

```python
n = 1024
x = random.normal(0, 1, n)
y = random.normal(0, 1, n)

scatter(x, y, s=75, c=arctan2(y, x), alpha=.5)

xticks([]); yticks([])
xlim(-1.5, 1.5); ylim(-1.5, 1.5);
```

$\S$ Exercise: DNA sequence visualisation (1/2)
===

We consider a DNA sequence as a walk in three-dimensional space:

* Each base (A, T, C, G) is assigned a direction in this space.
* Starting from the origin (0, 0, 0), each nucleotide in the sequence is a step.

```python
base_vectors = {'A': array([ 1,  1,  1]),
                'T': array([-1, -1,  1]),
                'C': array([-1,  1, -1]),
                'G': array([ 1, -1, -1])}
subplot(1, 1, 1, projection='3d')
for base in base_vectors:
    x, y, z = array([zeros(3), base_vectors[base]]).T
    plot(x, y, z, label=base)
legend();
```
