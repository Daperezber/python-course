About matplotlib
===

What is matplotlib?
---

* 2D plotting library for Python.
* Produces publication quality figures.
* Integrates with IPython and NumPy (in PyLab mode).

About matplotlib
===

Interfaces
---

### 1) Object-oriented interface

* Best for larger developments.
* You create and keep track of figures and axes.

### 2) State-machine environment (`pyplot`)

* For interactive plotting.
* matplotlib creates figures and axes for you.
* matplotlib keeps track of the current figure and axes.


About matplotlib
===

Interfaces
---

### 3) PyLab mode

* MATLAB-like environment based on IPython, NumPy, and matplotlib.
* Uses state-machine environment (`pyplot`).
* Imports everything from `matplotlib.pyplot` and `numpy` for you.
* Calls `show()` for you after plotting.

This is what we will be using initially.

About matplotlib
===

Interfaces: an example
---

1. Object-oriented interface:

    ```markdown
    import matplotlib.pyplot as plt
    import numpy as np
    x = np.arange(0, 10, 0.2)
    y = np.sin(x)
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot(x, y)
    ```

2. State-machine environment (`pyplot`):

    ```markdown
    import matplotlib.pyplot as plt
    import numpy as np
    x = np.arange(0, 10, 0.2)
    y = np.sin(x)
    plt.plot(x, y)
    ```

3. PyLab mode:

    ```markdown
    %pylab
    x = arange(0, 10, 0.2)
    y = sin(x)
    plot(x, y)
    ```

Step by step
===

A simple plot
---

We plot cosine and sine from $-\pi$ to $\pi$.

````python
x = linspace(-pi, pi, 256)
c, s = cos(x), sin(x)
plot(x, c)
plot(x, s);
````

We're going to progressively enhance this.

````python
# In this section, we want to continue working on the same figure.
%config InlineBackend.close_figures=False
````

Step by step
===

Changing colors and line widths
---

We want blue and red, a thicker line, and a more horizontal picture.

````python
figsize(10, 4)
plot(x, c, color='blue', linewidth=2.5, linestyle='-')
plot(x, s, color='red', linewidth=2.5, linestyle='-');
````

Step by step
===

Setting limits
---

The y-axis was a bit too tight.

````python
xlim(x.min() * 1.1, x.max() * 1.1)
ylim(c.min() * 1.1, c.max() * 1.1);
````

Step by step
===

Setting ticks
---

Ticks on the x-axis do not show the interesting values.

````python
xticks([-pi, -pi/2, 0, pi/2, pi])
yticks([-1, 0, +1]);
````

Step by step
===

Setting tick labels
---

* We could guess that 3.142 is $\pi$ but it would be better to make it explicit.
* When setting the ticks, we can also supply labels.
* The labels can contain LaTeX math mode.

````python
xticks([-pi, -pi/2, 0, pi/2, pi],
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1, 0, +1],
       [r'$-1$', r'$0$', r'$+1$']);
````

**Note:** In raw strings, we don't have to escape backslashes:

````python
r'Here we have a slash followed by the letter n: \n'
````

Step by step
===

Moving spines
---

We move the bottom and left spines to the middle and make the others invisible.

````python
ax = gca()  # gca stands for 'get current axis'
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position('zero')
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position('zero')

xticks([-pi, -pi/2, pi/2, pi],
       [r'$-\pi$', r'$-\pi/2$', r'$+\pi/2$', r'$+\pi$'])
yticks([-1, +1],
       [r'$-1$', r'$+1$']);
````

Step by step
===

Adding a legend
---

* Add `label` keyword arguments to `plot()`.
* These will be used in the legend box.

````python
# We are re-plotting (with label argument), so we first
# want to remove the existing lines.
#gca().lines.pop()
for l in gca().get_lines():
    gca().lines.remove(l)
````

````python
plot(x, c, color='blue', linewidth=2.5, linestyle='-', label='cosine')
plot(x, s, color='red', linewidth=2.5, linestyle='-', label='sine')
legend(loc='upper left');
````

Step by step
===

Devil is in the details
---

* Tick labels were hardly visible.
* We make them bigger.
* And render them on a semi-transparent white background.

````python
ax = gca()
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='none', alpha=0.65))
show()
````

Figures, subplots, axes and ticks
===

Figures
---

What PyLab's `plot()` really does:

* Call `gcf()` to get current figure, and `figure()` if there is none.
* Call `gca()` to get current axes, and `subplot(1, 1, 1)` if there is none.

A figure is the window to draw on and can contain multiple plots or axes.

* In a GUI, figures can be closed by clicking the `x` in the upper right corner.
* Programmatically, use the `close()` function.

Figures, subplots, axes and ticks
===

Subplots
---

To arrange plots in a regular grid we call `subplot()` with:

* The number of rows and columns.
* The subplot number (counted from 1, left to right, top to bottom).

````python
for i in range(1, 5):
    subplot(2, 2, i)
    xticks([]); yticks([])
    text(0.5, 0.5, 'subplot(2, 2, %d)' % i, ha='center', size=20, alpha=.5)
````

Figures, subplots, axes and ticks
===

Axes
---

Axes are similar to subplots, but can be placed at any location in the figure.

````python
axes([.1, .1, .8, .8])
xticks([]); yticks([])
text(.6, .6, 'axes([0.1, 0.1, .8, .8])', ha='center', size=20, alpha=.5)

axes([.2, .2, .3, .3])
xticks([]); yticks([])
text(.5, .5, 'axes([0.2, 0.2, .3, .3])', ha='center', size=16, alpha=.5);
````
