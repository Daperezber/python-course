$\S$ Exercise: Prime number sieve (solution)
===

* Use `ones` to initialize a boolean array with `True` values.
* Assign to all higher multiples of `i` by slicing with `[2*i::i]`.

```python
is_prime = np.ones((100,), dtype=bool)
is_prime[:2] = False

max_factor = int(np.sqrt(len(is_prime)))
for i in range(2, max_factor):
    is_prime[2*i::i] = False

np.nonzero(is_prime)
```
